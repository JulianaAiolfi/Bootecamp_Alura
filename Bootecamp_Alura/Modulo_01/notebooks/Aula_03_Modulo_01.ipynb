##Desafio 02: Adicionar uma coluna mostrando a região (Norte, Nodeste, Sul, Suldeste e Centro-Oeste) de cada estado.

```python
usaveis['Região'] = usaveis.index.str[0]

#Criando um dicionário para efetuar a substituição do número por um nome
regioes = {
'1': 'Norte',
'2': 'Nordeste',
'3': 'Sudeste',
'4': 'Sul',
'5': 'Centro-Oeste'
}
#alterando os dados da coluna criada para que constem os nomes das regiões e não um número
usaveis['Região'] = usaveis['Região'].replace(regioes)
```

```python
usaveis.head()
```

##Desafio 03: deixar o gráfico mais bonito. A mensagem mais clara

```python
ax = ordenado_por_total.head(5).T.plot(figsize=(10,6))
ax.legend(loc='best', bbox_to_anchor=(1, 1))

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.0f}")) #Editando a formatação do eixo Y
ax.set_ylabel("Valores em milhões de reais")

plt.title("Gasto total referente a internações no SIH/SUS dos cinco Estados Brasileiros que mais gastaram", fontsize=16) #Definindo título
ax.set_xticklabels( ax.get_xticklabels(), rotation=45, ha = "right") 
# rotacionando a legenda do Eixo X com a origem da rotação pela direita, get_xticklabels retorna os rótulos do eixo x

plt.show() # mostrando o gráfico
```
