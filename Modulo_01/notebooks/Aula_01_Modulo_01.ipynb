#Modulo_01 (Aula_01 Aquisição e leitura de dados reais)

Nesta aula vamos aprender sobre um conjunto de dados do governo brasileiro, da saúde extamente sobreos gastos de ambiente hospitalar, no SUS a partir de 2008


A base de dados foi selecionana dentro do Datasus no Tabnet, um sistema de acesso aos bancos de dados.


```python
import pandas as pd
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/JulianaAiolfi/Bootecamp_Alura/main/Modulo_01/dados/SUS_processamento_valor_anomes_unidfederacao.csv',
                    encoding='ISO-8859-1',skiprows=3, sep=';', skipfooter=12,
                    thousands='.', decimal=',')
dados.head()
```

```python
#Visualizando as últimas linhas

dados.tail()
```

```python
dados.mean()
```

```python
dados.info()
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
dados.mean()
```

```python
#Selecionando apenas uma única coluna

dados['2008/Ago']
```

```python
dados['2008/Ago'].mean()
```

```python
dados.plot(x='Unidade da Federação',y='2008/Ago')
```

```python
dados.plot(x='Unidade da Federação',y='2008/Ago', kind='bar', figsize=(10,7))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
ax = dados.plot(x='Unidade da Federação',y='2008/Ago', kind='bar', figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.title('Valor por unidade da Federação')
plt.show()
```

# Desafio_01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.

```python
ax = dados.plot(x='Unidade da Federação',y='2008/Ago', kind='bar', figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.title('SUS - Valor (R$) processado por unidade da Federação em Agosto/2008')
plt.show()
```

# Desafio_02: Faça a mesma análise realizada em aula, porém para o mês mais recente.

```python
ax = dados.plot(x='Unidade da Federação',y='2021/Jul', kind='bar', figsize=(10,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.title('SUS - Valor(R$) processado por unidade da Federação em Julho/2021')
plt.show()
```
