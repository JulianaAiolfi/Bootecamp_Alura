# Aula_02 Primeiras visualizações de dados

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker

import warnings
warnings.filterwarnings('ignore')
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/JulianaAiolfi/Bootecamp_Alura/main/Modulo_01/dados/SUS_atendimento_valor_anomes_unidfederacao.csv',
                    encoding='ISO-8859-1',skiprows=3, sep=';', skipfooter=12,
                    thousands='.', decimal=',')
```

```python
pd.options.display.float_format = '{:.2f}'.format

dados.head()
```

```python
dados[['2008/Ago','2008/Set']].head()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, 'Unidade da Federação')
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
usaveis = usaveis.set_index('Unidade da Federação')
usaveis.head()
```

```python
usaveis['2019/Ago'].head()
```

```python
usaveis.loc['12 Acre']
```

```python
usaveis.plot(figsize=(12,6))
```

Descrição gráfico
- Título: Sem título
- Base de dados utilizada: Variável 'usaveis'
- Tipo de gráfico: Linha
- Eixo X: Estado
- Eixo Y: Valores de todos os meses de ago/2007 até jul/2021 e o total 
- O que está sendo mostrado: Foi criado um gráfico de linha dos 26 Estados Brasileiros mais o Distrito Federal relacionando ao seu custo por internação referente ao mês de agosto de 2007 até julho de 2020, retirando as colunas que não possuiam dados preenchidos. É possível observar uma legenda que sobrepos o gráfico e conta com todos os meses analisados. 

```python
# Utilizamos o "T" para "girar" ou transpor, o DataFrame, cada linha se torna uma coluna

usaveis.T.head()
```

```python
usaveis.T.plot(figsize=(10,6))
```

Descrição gráfico
- Título: Sem título
- Base de dados utilizada: Variável 'usaveis'
- Tipo de gráfico: Linha
- Eixo X: Anos/meses
- Eixo Y: Valores
- O que está sendo mostrado: Foi feita a transposição do DataFrame usaveis e gerado o gráfico com os vinte e seis Estados Brasileiros e Distrito Federal aparecendo agora na legenda. 

```python
# O drop vai remover uma coluna ou linha do DataFrame baseado no parâmetro axis
# com axis=1 estamos removendo a última coluna

usaveis = usaveis.drop('Total', axis=1)
usaveis.head()
```

```python
usaveis.T.plot(figsize=(10,6))
```

Descrição gráfico
- Título: Sem título
- Base de dados utilizada: Variável 'usaveis'
- Tipo de gráfico: Linha
- Eixo X: Anos/meses
- Eixo Y: Valores
- O que está sendo mostrado: Foi feito o drop da coluna total e a visualizaçào do gráfico foi alterada gerando um novo resultado em que podemos observar uma sazonalidade e crescimento. Contendo na legenda os vinte e seis estados brasileiros e o Distrito Federal.

# Desafio_01: Reposicionar a legenda do gráfico em uma posição mais adequada

```python
# O .plot retona um objeto que representa os eixos do gráfico
# aqui salvamos esse objeto em ax para futuras configurações
ax = usaveis.T.plot(figsize=(10,6))


# A partir do ax configuramos a posição da legenda com o legend
ax.legend(loc='upper center', bbox_to_anchor=(0.5,-0.07), ncol=4) # upper center posicionar no meio e o bbox_to_anchor ajustar posição e ncol = número de colunas
```

# Desafio_02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```python
usaveis[:5].T.plot(figsize=(10,6))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
