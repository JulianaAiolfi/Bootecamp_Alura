# Aula_02 Primeiras visualizações de dados

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
dados = pd.read_csv('https://raw.githubusercontent.com/JulianaAiolfi/Bootecamp_Alura/main/Modulo_01/dados/SUS_atendimento_valor_anomes_unidfederacao.csv',
                    encoding='ISO-8859-1',skiprows=3, sep=';', skipfooter=12,
                    thousands='.', decimal=',')
```

```python
pd.options.display.float_format = '{:.2f}'.format

dados.head()
```

```python
dados[['2008/Ago','2008/Set']].head()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, 'Unidade da Federação')
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
usaveis = usaveis.set_index('Unidade da Federação')
usaveis.head()
```

```python
usaveis['2019/Ago'].head()
```

```python
usaveis.loc['12 Acre']
```

```python
usaveis.plot(figsize=(12,6))
```

```python
usaveis.T.head()
```

```python
usaveis.T.plot(figsize=(10,6))
```

```python
usaveis = usaveis.drop('Total', axis=1)
usaveis.head()
```

```python
usaveis.T.plot(figsize=(10,6))
```
